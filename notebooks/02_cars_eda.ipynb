```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("../data/cars_versions_specs.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Numerical columns
numerical_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()

print("Numerical columns:", numerical_cols)
```

```python
# Categorical columns
categorical_cols = df.select_dtypes(include=['object','category']).columns.tolist()

print("Categorical columns:", categorical_cols)
```

```python
counts = {'Numerical': len(numerical_cols), 'Categorical': len(categorical_cols)}

plt.figure(figsize=(6,4))
plt.bar(counts.keys(), counts.values(), color=['skyblue','salmon'])
plt.title("Number of Numerical vs Categorical Columns")
plt.show()
```

# Histograms

```python
# Price histogram 
sns.histplot(df['price'], bins=20, kde=False) 
plt.title("price Distribution") 
plt.show()
```

```python
# motorisation_puissance_fiscale histogram

plt.figure(figsize=(12, 6))
sns.histplot(df['motorisation_puissance_fiscale'], bins=20, kde=False) 
plt.title("motorisation_puissance_fiscale Distribution") 
plt.show()

```

```python
# motorisation_puissance_(ch.din) histogram
sns.histplot(df['motorisation_puissance_(ch.din)'], bins=20, kde=False) 
plt.title("motorisation_puissance_(ch.din) Distribution") 
plt.show()
```

```python
# Energy type countplot
plt.figure(figsize=(13,6))
sns.countplot(x='motorisation_energie', data=df, palette="Set2")
plt.title("Energy Type Count")
plt.xticks(rotation=45)  # rotate labels if they are long
plt.show()
```

# Missing Values

```python
missing_counts = df.isnull().sum()

plt.figure(figsize=(10,6))
missing_counts[missing_counts > 0].sort_values().plot.barh(color="pink")
plt.title("Missing Values per Column")
plt.xlabel("Count of Missing Values")
plt.show()
```

# Correlation

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10, 8)) 
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5) 
plt.title("Correlation Heatmap") 
plt.show()
```

# Scatterplot


```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x=df['motorisation_puissance_(ch.din)'], 
    y=df['price']
)
plt.title("Power (ch DIN) vs Price")
plt.xlabel("Power (ch DIN)")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x=df['motorisation_puissance_fiscale'], 
    y=df['price']
)
plt.title("Power vs Price")
plt.xlabel("Power")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x=df['motorisation_puissance_fiscale'], 
    y=df['motorisation_puissance_(ch.din)']
)
plt.title("Power vs Power(ch.din)")
plt.xlabel("Power")
plt.ylabel("Power(ch.din)")
plt.show()
```

# Outliers 

```python
# Create subplots: 1 row, 3 columns
fig, axes = plt.subplots(1, 3, figsize=(20, 10))

# --- Numeric variables (boxplots) ---
sns.boxplot(y=df['price'], ax=axes[0], color="skyblue")
axes[0].set_title("price Outliers")

sns.boxplot(y=df['motorisation_puissance_fiscale'], ax=axes[1], color="lightgreen")
axes[1].set_title("motorisation_puissance_fiscale Outliers")

sns.boxplot(y=df['motorisation_puissance_(ch.din)'], ax=axes[2], color="salmon")
axes[2].set_title("motorisation_puissance_(ch.din) Outliers")

# Adjust layout
plt.tight_layout()
plt.show()

```
