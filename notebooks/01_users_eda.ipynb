```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("../data/users_dataset.csv")
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Numerical columns
numerical_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()

print("Numerical columns:", numerical_cols)
```

```python
# Categorical columns
categorical_cols = df.select_dtypes(include=['object','category']).columns.tolist()

print("Categorical columns:", categorical_cols)
```

```python
counts = {'Numerical': len(numerical_cols), 'Categorical': len(categorical_cols)}

plt.figure(figsize=(6,4))
plt.bar(counts.keys(), counts.values(), color=['skyblue','salmon'])
plt.title("Number of Numerical vs Categorical Columns")
plt.show()
```

# Histograms

```python
# Age histogram 
sns.histplot(df['age'], bins=20, kde=False) 
plt.title("Age Distribution") 
plt.show()
```

```python
# Budget histogram
sns.histplot(df['budget_max'], bins=20, kde=False) 
plt.title("Budget Distribution") 
plt.show()
```

```python
# Revenue histogram
sns.histplot(df['revenu_mensuel'], bins=20, kde=False) 
plt.title("Revenue Distribution") 
plt.show()
```

```python
# Carrosserie countplot 
plt.figure(figsize=(12, 6))
sns.countplot(x='carrosserie_preferee', data=df, palette="Set2") 
plt.title("Carrosserie Count") 
plt.show()
```

```python
# Energy preference countplot 
sns.countplot(x='energie_preferee', data=df, palette="Set2") 
plt.title("Energy Preference Count")
plt.show()
```

# Correlation

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10, 8)) 
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5) 
plt.title("Correlation Heatmap") 
plt.show()
```

# Scatterplot

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['budget_max'], y=df['revenu_mensuel'], hue=df['revenu_mensuel'], palette="coolwarm")
plt.title("Budget vs Revenue")
plt.show()
```

# Outliers 

```python
# Create subplots: 1 row, 3 columns
fig, axes = plt.subplots(1, 3, figsize=(20, 10))

# --- Numeric variables (boxplots) ---
sns.boxplot(y=df['revenu_mensuel'], ax=axes[0], color="skyblue")
axes[0].set_title("Revenue Outliers")

sns.boxplot(y=df['budget_max'], ax=axes[1], color="lightgreen")
axes[1].set_title("Budget Outliers")

sns.boxplot(y=df['age'], ax=axes[2], color="salmon")
axes[2].set_title("Age Outliers")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
sns.histplot(df['age'], bins=30, kde=True) 
plt.title("Age Distribution with Outliers") 
plt.show()
```
